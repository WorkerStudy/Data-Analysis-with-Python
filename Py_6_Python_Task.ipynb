## Задача 1

Аня и Наташа играют в города. Они очень любят эту игру, знают много городов и к концу игры забывают, какие уже называли. На вас возложена почётная задача вести запись игры и напоминать девочкам, если какой-то город уже был назван.

**Формат ввода**

В первой строке записано число названных городов N. Затем идут N строк с названиями городов и ещё одна строка с новым только что названым городом.

**Формат вывода**

Слово OK, если такого города ещё не было названо, и TRY ANOTHER, если город уже был назван.

**Пример 1**

**Ввод**

3

Москва

Нью-Йорк

Лондон

Париж

**Вывод**

OK

**Пример 2**

**Ввод**

8

Лондон

Париж

Москва

Вашингтон

Берлин

Вена

Мадрид

Рим

Вена

**Вывод**

TRY ANOTHER

```python
num = int(input())
listCity= list()
while num>0:
    listCity.append(input())
    num-=1

newCity = input()

if newCity in listCity:
    print("TRY ANOTHER")
else:
    print("NO")
```

```python
numMed= int(input())
listHerbs= list()

while numMed > 0:
    numHerbs= int(input())
    while numHerbs > 0:
        herb= input()
        if herb  not in listHerbs:
            listHerbs.append(herb)
        numHerbs-=1
    numMed-=1

for word in listHerbs:
    print(word,sep="\n")
```

**Пример**
